```python
#эллипс
x0 = 2
y0 = 2
a = 4
b = 2
x = np.linspace(-4, 4, 200)
y1 = b/a * np.sqrt((a**2 - (x)**2))
y2 = - b/a * np.sqrt((a**2 - (x)**2))
plt.ylim(-5, 5)
plt.xlim(-5, 5)
plt.grid(True)
plt.plot(x, y1)
plt.plot(x, y2)
```

```python
#специальная функция
def polar2cart(r,phi):
    x = r*math.cos(phi)
    y = r*math.sin(phi)
    return x,y

polar2cart(5,math.pi/4)
```

Задание №4

1) Решить систему уравнений

```python
x = np.linspace(-5, 5, 200)
y1 = x**2 - 1
y2 = 1 - 1/x + np.exp(x)/x
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt,show()
```

```python
def equations(p):
    x, y = p
    return(x**2 - 1 - y, np.exp(x) + x * (1 - y) - 1)
x1, y1 = fsolve(equations, (-1,1))
print(x1, y1)
x2, y2 = fsolve(equations, (3,6))
print(x2, y2)
x3, y3 = fsolve(equations, (4,15))
print(x3, y3)
```
